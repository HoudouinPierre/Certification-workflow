# Parameters

```python
# Simple information structure
dimension                             = 5
information_structure_bound           = 1

# Properly scaled scenarios
properly_scaled_scenarios_max_L2_norm = 2

# Black-box function
return_max                            = False
normalization_quantile_level          = 0.9
N_empirical_distribution              = 1000
verbose                               = True
```

```python
information_structure    = SimpleInformationStructure(dimension, information_structure_bound)
properly_scaled_scenario = SimpleProperlyScaledScenario(properly_scaled_scenarios_max_L2_norm)
sampler                  = SimpleSampler(information_structure, properly_scaled_scenario)
```

# 1 - Linear black-box function

```python
black_box_function_name = "Linear black-box function"
```

```python
linear_simple_black_box_function = SimpleBlackBoxFunction(black_box_function_name, return_max, sampler, normalization_quantile_level, N_empirical_distribution, verbose)
linear_simple_black_box_function.compute_normalization_quantile()
N = 10
for i in range(N):
    scenario = sampler                         .draw_properly_scaled_scenario()
    y        = linear_simple_black_box_function.simulation(scenario)
    print("Scenario %d/%d: " % (i + 1, N) + str(scenario) + " - Output: " + str(y) + "     " + str(type(y)))
```
